```python
import number_theory
from functools import reduce

print(number_theory.gcd(750, 300))
print(number_theory.lcm(750, 300))

print(number_theory.get_euklid_elements(1433, 4060))
print(number_theory.inv_mod(1433, 4060))

print(private_key := number_theory.find_private_key((4189, 1433)))

encrypted_message = [1200, 3900]
decrypted_message = number_theory.decrypt_encrypt(encrypted_message, private_key)

# Letter to number conversion
letter_to_value = {chr(num + 96): num for num in range(1, 27)}
letter_to_value.update({chr(num + 34): num for num in range(31, 57)})
letter_to_value[""], letter_to_value[""], letter_to_value[""] = 27, 28, 29
letter_to_value[""], letter_to_value[""], letter_to_value[""] = 57, 58, 59
letter_to_value[" "] = 30
letter_to_value["."], letter_to_value[","], letter_to_value["!"] = 60, 61, 62

value_to_letter = {value: key for key, value in letter_to_value.items()}

split_message = reduce(lambda a, b: a + b, ((item // 64, item % 64) for item in decrypted_message))
message_str = "".join(value_to_letter[letter] for letter in split_message)
print(message_str)
```
